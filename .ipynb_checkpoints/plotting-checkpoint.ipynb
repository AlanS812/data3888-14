```python
import pandas as pd
xg = pd.read_csv('xgboost_augmented_metrics.csv')
rn = pd.read_csv('resnet50_augmented_metrics.csv')
rn.head()
```

```python
# prompt: Using dataframe xg: line plot of accuracy on y axis and blur on x axis, then noise as a categorical variable for colour. save the figure high res. use griddo not use altair

import matplotlib.pyplot as plt
import seaborn as sns

# Create the line plot using seaborn with a different palette
plt.figure(figsize=(10, 6))
sns.lineplot(data=xg, x='blur', y='accuracy', hue='noise', marker='o', palette='viridis')

# Add grid
plt.grid(True)

# Set title and labels
plt.title('Accuracy vs Blur for Different Noise Levels - XGBoost')
plt.xlabel('Blur')
plt.ylabel('Accuracy')

# Save the figure in high resolution
plt.savefig('accuracy_vs_blur_noise.png', dpi=300)

# Show the plot
plt.show()
```

```python
# Create the line plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=rn, x='blur_size', y='accuracy', hue='noise_level', marker='o', palette='viridis')

# Add grid
plt.grid(True)

# Set title and labels
plt.title('Accuracy vs Blur for Different Noise Levels - ResNet50')
plt.xlabel('Blur')
plt.ylabel('Accuracy')

# Save the figure in high resolution
plt.savefig('accuracy_vs_blur.png', dpi=300)

# Show the plot
plt.show()
```

```python
# prompt: add noise on 'cell_67_100.png

from PIL import Image
import numpy as np

# Load the image
img_path = 'cell_67_100.png'
img = Image.open(img_path).convert('RGB')
img_array = np.array(img)

# Define parameters for adding Gaussian noise
mean = 0
std_dev = 20 # Adjust standard deviation to control noise intensity

# Create Gaussian noise
noise = np.random.normal(mean, std_dev, img_array.shape).astype('uint8')

# Add noise to the image array
noisy_img_array = img_array + noise

# Clip the pixel values to be within the valid range [0, 255]
noisy_img_array = np.clip(noisy_img_array, 0, 255)

# Convert the noisy image array back to a PIL Image
noisy_img = Image.fromarray(noisy_img_array.astype('uint8'))

# Save the noisy image
noisy_img_path = 'cell_67_100_noisy.png'
noisy_img.save(noisy_img_path)

print(f"Noisy image saved to {noisy_img_path}")
```
